# 2. Data analysis

```python
import pandas as pd

data = []  # 데이터를 저장할 리스트

for gsm_name, gsm in gse.gsms.items():
    description = gsm.metadata['description']
    characteristics = gsm.metadata['characteristics_ch1']
    
    # gender와 condition 추출
    gender = ""
    condition = ""
    for char in characteristics:
        if 'female' in str(char).lower():
            gender = 'female'
        elif 'male' in str(char).lower():
            gender = 'male'
    for desc in description:
        if 'normal' in str(desc).lower():
            condition = 'control'
        elif "disease" in str(desc).lower():
            condition = "AD"
    
    data.append([gender, condition])  # 데이터 리스트에 추가

# 데이터프레임 생성
metadata = pd.DataFrame(data, columns=['Gender', 'Condition'], index=gse.gsms.keys())

# 생성된 데이터프레임 출력
metadata
```

```python
# Extract control / AD samples name
control_samples = metadata[metadata['Condition'] == 'control'].index
AD_samples = metadata[metadata['Condition'] == 'AD'].index
```

```python
# Extract Female / Male samples name
Female_samples = metadata[metadata['Gender'] == 'female'].index
Male_samples = metadata[metadata['Gender'] == 'male'].index
```

```python
# Expr_data = sample별 유전자 발현에 대한 정보
Expr_data = gse.pivot_samples('VALUE')
Expr_data
```

```python
control_expr_data = Expr_data[control_samples]
control_expr_data
```

```python
AD_expr_data = Expr_data[AD_samples]
AD_expr_data
```

## 2-1. Filtering : Extract probe above threshold

```python
Expr_samples_average = Expr_data.median(axis=1)
print("Number of probes before filtering: ", len(Expr_samples_average))
```

```python
# Quantile normalization 함수 정의
def quantile_normalization(df):
    rank_mean = df.stack().groupby(df.rank(method='first').stack().astype(int)).mean()
    return df.rank(method='min').stack().astype(int).map(rank_mean).unstack()
```

```python
# Quantile normalization 적용
normalized_expr = quantile_normalization(Expr_data)
normalized_expr
```

```python
# filtering 이전의 전체 index 개수 확인
expr_threshold = Expr_samples_average.quantile(0.05)
expressed_probes = Expr_samples_average[Expr_samples_average >= expr_threshold].index.tolist()
print("Number of probes above threshold: ", len(expressed_probes))
```

```python
# threshold로 사용할 인덱스 추출
top5_percent_index = normalized_expr > normalized_expr.quantile(0.95,axis=0)
top5_percent_index.iloc[:,1].value_counts()
```

```python
# 모든 샘플의 절반 이상에서 선택된 인덱스 추출
selected_index = top5_percent_index[top5_percent_index.sum(axis=1) >= len(normalized_expr.columns) / 2].index.tolist()
selected_index_list = []  

for si in selected_index :
    selected_index_list.append(si)
    
print("Number of probes above threshold: ", len(selected_index_list))
```

```python
# 추출된 probe로 구성된 새로운 expr_matrix 구성
new_expr_data = Expr_data[np.isin(Expr_data.index, selected_index_list)]
# transpose data
new_expr_data = np.transpose(new_expr_data)
new_expr_data
```

#### new_expr_data를 성별에 따라 나눔

```python
female_expr_data = new_expr_data.loc[Female_samples, :]
female_expr_data
```

```python
male_expr_data = new_expr_data.loc[Male_samples, :]
male_expr_data
```

**Visualization : 차원 축소를 통한 시각화**

```python
from sklearn.decomposition import PCA

# 함수 선언. PCA 적용할 데이터를 인자로 받아서 PCA fitting 결과를 되돌려 줌
def dimensionality_reduction(X, n_components, data):
    #객체 선언
    pca = PCA(n_components=n_components)

    #적용
    pca.fit(X)
    x_pca = pca.transform(X)

    #x_pca를 보기 쉽게 데이터프레임으로 만들기
    pca_df = pd.DataFrame(x_pca, columns=['PC'+str(i) for i in range(1, n_components+1)], index=list(data.index.values))
    
    return pca_df, pca
```

```python
# 함수 호출
pca_df, x_pca = dimensionality_reduction(X_df, 2, data)

# cluster 결과를 붙임
pca_df['cluster'] = km_cluster
print(pca_df)

# 누적 설명 분산량
# 설정한 주성분의 개수(n_components)로 전체 데이터의 분산을 얼마나 설명할 수 있는가를 나타냄
# 언더바(_)는 분석이 진행된 이후의 결과 값을 나타냄
print(x_pca.explained_variance_ratio_, sum(x_pca.explained_variance_ratio_))
```

```python
# condition 열 추가
pca_df['condition'] = ['control' if index in control_samples else 'AD' for index in pca_df.index]
pca_df
```

```python
# 시각화를 위한 함수 선언
import matplotlib.pyplot as plt
from itertools import product
def plot_clustering_result(pca_df, cluster_labels:list, labels:list = ['control', 'AD'], title='Total_Sample_PCA'):
    assert len(labels) == 2
    assert len(cluster_labels) == 2
    plt.style.use('default')
    
    fig = plt.figure(figsize = (8, 8))
    ax = fig.add_subplot(1, 1, 1)
    ax.set_xlabel('PC1', fontsize = 15)
    ax.set_ylabel('PC2', fontsize = 15)
    ax.set_title(title, fontsize=20)

    colors = ['red', 'salmon', 'blue', 'dodgerblue']
    markers = ['o', 's']

    cluster0_c = pca_df[(pca_df['cluster'] == cluster_labels[0]) & (pca_df['condition'] == labels[0])]
    cluster0_A = pca_df[(pca_df['cluster'] == cluster_labels[0]) & (pca_df['condition'] == labels[1])]
    cluster1_c = pca_df[(pca_df['cluster'] == cluster_labels[1]) & (pca_df['condition'] == labels[0])]
    cluster1_A = pca_df[(pca_df['cluster'] == cluster_labels[1]) & (pca_df['condition'] == labels[1])]

    scatter_label = []
    for i, j in product([0, 1], repeat=2):
        scatter_label.append('cluster-'+str(cluster_labels[i])[:2]+' & '+labels[j])

    ax.scatter(cluster0_c['PC1'], cluster0_c['PC2'], c=colors[0], marker=markers[0], label=scatter_label[0])
    ax.scatter(cluster0_A['PC1'], cluster0_A['PC2'], c=colors[1], marker=markers[1], label=scatter_label[1])
    ax.scatter(cluster1_c['PC1'], cluster1_c['PC2'], c=colors[2], marker=markers[0], label=scatter_label[2])
    ax.scatter(cluster1_A['PC1'], cluster1_A['PC2'], c=colors[3], marker=markers[1], label=scatter_label[3])

    ax.legend()
    ax.grid()
    
    plt.show()
```

```python
plot_clustering_result(pca_df=pca_df, labels=['control', 'AD'], cluster_labels=[0, 1])
```

### Choosing the appropriate number of clusters

* Two common methods that are commonly used to evaluate the appropriate number of clusters:
1) The elbow method
2) The silhouette coefficient

```python
# Elbow methods
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42,
}

# A list holds the SSE values for each k

# Female data
sse = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(data_scale_f)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), sse)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

### 2-3-2 Naive-bayesian classification

### (1) Total sample classification

```python
X_df_2 = (X_df).join(metadata['Condition'], how='left')
X_df_2
```

```python
# GaussianNB 객체 생성
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB().fit(X_train, y_train)
```

```python
# 예측
y_test_pred = clf.predict(X_test)
print(y_test_pred[:3]) # 예측한 결과 샘플 3개 출력
print(clf.get_params()) # GaussianNB 클래스 인자 설정 정보
print('정확도 : ', clf.score(X, y)) # 성능 평가 점수(Accuracy)
```

```python
print("Number of mislabeled points out of a total %d points : %d"
%(X_test.shape[0], (y_test != y_test_pred).sum()))
```

**Visualization : plot confusion matrix**

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

# make confusion matrix
cm = confusion_matrix(y_test, y_test_pred, labels=['control', 'AD'])

# plot confusion matrix
plt.style.use("default")
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['control', 'AD'])
disp.plot()
plt.show()
```
